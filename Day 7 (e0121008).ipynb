```r
# sum of n natural numbers 
a=as.integer(readline("enter the number "))
b=0
while(a>0)
    {
    b=b+a
    a=a-1
}
print(b)
```

```r
#factorial of given number
a=as.integer(readline("enter the number "))
b=1
while(a>0)
    {
    b=b*a
    a=a-1
}
print(b)
```

```r
# get the  subject detail print in sorted
# get input from user for a list
library(stringr)
element=c()
total=as.integer(readline("enter the number of subject:"))  
for(i in seq(total))
    {
        element[i]=(readline(paste("enter subject ",i,":")))
}
paste(str_sort(element))
#print(sort(element))
```
