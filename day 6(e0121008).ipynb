```r
#Reading and writing execl file
library("xlsx")
#Reading the file
df=read.xlsx("Titanic-clean.xlsx",sheetName="New_Titanic")#sheet name to select sheets  with help of sheet name excel
df
df=read.xlsx("Titanic-clean.xlsx",sheetIndex=1)#sheet index to select sheets in excel
df
```

```r
#Creating the data frame
df2=data.frame(name=c("William","Emma","Sofia","Markus","Edward"),region=factor(c("East","North","East","South","West")),sales=c(50000,52000,90000,34000,42000),expense=c(42000,43000,50000,44000,38000))
df2
```

```r
#To write a df in excel
write.xlsx(df2,file="AIMLWRITE.xlsx") # write used to convert data frame into excel
```

```r
#To write a df in excel
write.xlsx(df2,file="AIMLWRITE.xlsx",sheetName="sample")
write.xlsx(df2,file="AIMLWRITE.xlsx",sheetName="sample",row.name=FALSE)#To delete the default row name
```
