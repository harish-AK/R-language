```r
# MODULE 3
# DATA VISUALIZATION -GGPLOT2 USED TO VISUALIZE IN R

df=data.frame(dose=c('D0.5','D1','D2'),len=c(4.2,10,29.5))
head(df)
```

```r
install.packages('ggplot2')
library('ggplot2')
```

```r
library('ggplot2')
```

```r
# basic bar plot 
p=ggplot(data=df,aes(x=dose,y=len))+geom_bar(stat='identity')
p

# horizontal bar plot
p+coord_flip()  # coord flip used to print the graph in horizontally
```

```r
# basic bar plot 
p=ggplot(data=df,aes(x=dose,y=len))+geom_bar(stat='identity',width=0.3)   # width  0.3
p

# horizontal bar plot
p+coord_flip()
```

```r
# basic bar plot 
p=ggplot(data=df,aes(x=dose,y=len))+geom_bar(stat='identity',color='red',fill='violet',width=0.3) # color=outline(border) and fill=bar color
p

# horizontal bar plot
p+coord_flip()
```

```r
# basic bar plot 
p=ggplot(data=df,aes(x=dose,y=len))+geom_bar(stat='identity',fill='gold')+theme_minimal()  
p
```

```r
# basic bar plot 
p=ggplot(data=df,aes(x=dose,y=len))+geom_bar(stat='identity',fill='gold')+theme_classic()  # classic -remove the border
p
```

```r
# basic bar plot 
p=ggplot(data=df,aes(x=dose,y=len))+geom_bar(stat='identity',fill='gold')+theme_dark()  # theme_dark make the background as dark
p
```

```r
# basic bar plot 
p=ggplot(data=df,aes(x=dose,y=len))+geom_bar(stat='identity',fill='gold')+theme_test()  # remove the background grids
p
```

```r
# basic bar plot 
p=ggplot(data=df,aes(x=dose,y=len))+geom_bar(stat='identity',fill='gold')+theme_grey()  
p
```

```r
colors = c("green", "orange", "brown")    
subjects <- c("R", "Python", "Java")
marks <- c("Vargeese", "Pragdeesh", "Ramesh")
  
Values <- matrix(c( 9, 11, 4, 7, 12, 2, 8, 10, 11),                                                                    
                 nrow = 3, ncol = 3, byrow = TRUE)
  
barplot(Values, main = "Subject Marks", names.arg = subjects,  
                          xlab = "Subject", ylab = "Marks", 
                          col = colors, density = 100,angle=47,beside = TRUE)  #angle used to color with the specified angle
  # destiny used to thicken the colors
legend("topright", marks, cex =1, fill = colors) #cex=legends size
```

```r
colors = c("green", "orange", "brown")
months <- c("Mar", "Apr", "May", "Jun", "Jul")
regions <- c("East", "West", "North")
  
Values <- matrix(c(2, 9, 3, 11, 9, 4, 8, 7, 3, 12, 5, 2, 8, 10, 11), 
                 nrow = 3, ncol = 5, byrow = TRUE)
  
barplot(Values, main = "Total Revenue", names.arg = months,  xlab = "Month", ylab = "Revenue", col = colors, density = 10,beside = TRUE)
  
legend("topleft", regions, cex = 0.7, fill = colors)
```
